In this case I simply computed the number of weeks as (rounded) number of days / 7. Another approach could consider a 'week' as the n-th week of the year starting from january 1st:

```python
def number_of_weeks(year):
    date = pd.to_datetime(f"{year}/12/28") # december 28th is always in the last week of the year according to ISO 8601
    return date.week
```

```python
from tqdm import tqdm

def get_weeks(data1, data2):
    """
        Get difference of weeks between two dates
        data1: earliest data column (dogs.posted)
        data2: closest data column (dogs.accessed)
        Return: 
    """
    weeksdelta = []
    for i in tqdm(range(len(data1))):
        if data1[i].year == data2[i].year: # if dates are in the same year
            weeksdelta.append(data2[i].week-data1[i].week)
        else:
            nweeks = number_of_weeks(data1[i].year)
            delta = nweeks - data1[i].week # remaining weeks from latest data to end of year
            for year in range(data1[i].year + 1, (data2[i].year - 1)): # for each year in between
                delta += number_of_weeks(year)
            delta += data2[i].week
            weeksdelta.append(delta)
    return weeksdelta
```

### 12. Find for duplicates in the dogs dataset. Two records are duplicates if they have (1) same breeds and sex, and (2) they share at least 90% of the words in the description field. Extra points if you find and implement a more refined for determining if two rows are duplicates

To do so, first of all I preprocess the description fields in order to obtain lists of tokens. The preprocessing steps consist of:
1. Remove links, numbers, punctuation, double whitespaces; convert text to lowercase; separate sentences with dots;
2. Tokenize;
3. Lemmatize.

These tokens are a better representation of a text since they basically don't include the so called stopwords (articles, conjunctions, ecc.) and for example repeated verbs and words with same roots but different endings. This also permits to better perform vectorization approaches and in general to reduce computational times.

```python
import nltk
from nltk.corpus import stopwords, wordnet
from nltk.stem import WordNetLemmatizer

def remove_stopwords(text):
    stop_words = set(stopwords.words('english'))
    return ' '.join([word for word in text.split() if word not in stop_words])

def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

def lemmatize(text):
    lemmatizer = WordNetLemmatizer()
    tqdm.pandas(desc="Pos tagging")
    pos_tag = text.progress_apply(lambda x: nltk.pos_tag(x))
    wordnet_pos = pos_tag.progress_apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
    tqdm.pandas(desc="Lemmatization")
    lemmatized = wordnet_pos.progress_apply(lambda x: [lemmatizer.lemmatize(word, tag) for word, tag in x])
    return lemmatized

def preprocess_text(descr):
    # remove links
    descr = descr.str.replace('http\S+|www.\S+', '', case=False, regex=True)
    # lowercase
    descr = descr.str.lower()
    # remove numbers
    descr = descr.str.replace(r'\d+', ' ', regex=True)
    # separate with ". "
    descr = descr.str.replace(r'\.(?=[^ \W\d])', '. ', regex=True)
    # remove punctuation
    descr = descr.str.replace(r'[^\w\s]', '', regex=True)
    # remove double withespaces
    descr = descr.str.replace(r'\s+', ' ', regex=True)
    # remove withespaces at the start and end of a sentence
    descr = descr.apply(lambda x: x.strip())
    # remove stopwords
    tqdm.pandas(desc="Removing stopwords")
    descr = descr.progress_apply(lambda x: remove_stopwords(x))
    # tokenization
    tqdm.pandas(desc="Tokenization")
    tokens = descr.progress_apply(lambda x: x.split())
    # lemmatization
    descr = lemmatize(tokens)
    
    return descr

dogs['tokens'] = preprocess_text(dogs[dogs.description.isna()==False].description)
```

```python
def find_intersection(row_tokens, next_rows_tokens):
    common_tokens = [len(set(row_tokens).intersection(set(tokens))) for tokens in next_rows_tokens]

    return common_tokens


def compare_tokens(df):
    similarities = []
    df_dict = df.to_dict('index')
    for i, row in tqdm(df_dict.items()):
        next_rows = df.iloc[i+1:]
        
        # number of common tokens divided by sum of all unique tokens must be >= 0.9
        common_tokens = [len(set(row['tokens']).intersection(set(tokens))) for tokens in next_rows['tokens']]
        tot_tokens = [len(set(row['tokens']).union(set(tokens))) for tokens in next_rows['tokens']]
        similarity = [x/y for x,y in zip(common_tokens, tot_tokens)]
        similarities.extend(similarity)
        
    df.loc[1:, 'similarity'] = similarities # append column starting from 2nd row
        
    return df[df.similarity >= 0.9]


# duplicates = compare_tokens(dogs)

# keep duplicates based on sex and breed
# duplicates.duplicates(subset=['sex', 'breed_primary'], keep='last')
```

The code above takes too much time since it has to compare each row with all the following ones. Another method has been developed, which is based on a self join in order to iterate through all rows much faster.

```python
dogs = dogs.sort_values('id')
dogs_filt = dogs[['id','breed_primary','sex','tokens']]

# self join based on sex and breed
joined_df = pd.merge(dogs_filt, dogs_filt, on=['breed_primary','sex'])

# dropping comparisons between same id (id_x != id_y) and ensuring that there aren't repeated comparisons (id_x < id_y)
joined_df = joined_df[(joined_df['id_x'] != joined_df['id_y']) & (joined_df['id_x'] < joined_df['id_y'])].reset_index(drop=True)
```

```python
import time

def compare_tokens(tokens_x, tokens_y):
    iou = len(set(tokens_x).intersection(set(tokens_y))) / len(set(tokens_x).union(set(tokens_y)))
    return iou

start_time = time.time()
iou = []
for row in tqdm(joined_df.values, total=len(joined_df)): # numpy vectorization cause is one of the fastest methods
    try:
        iou.append(compare_tokens(row[3], row[5]))
    except ZeroDivisionError: # due to missing descriptions (so no tokens)
        iou.append(np.nan)
         
end_time = time.time()

joined_df['iou'] = iou


print(f"Execution time: {round(end_time-start_time)} seconds")
```

It's possible to refine this method with a vectorization approach using TF-IDF representation and cosine similarity which can capture better than IoU the meaning of a description.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.metrics.pairwise import cosine_similarity

dogs['clean_description'] = [" ".join(x) for x in dogs['tokens']]
docs = list(dogs['clean_description'])
vectorizer = TfidfVectorizer(min_df=0.005)
X = vectorizer.fit_transform(docs)
X.shape
```

```python
X = X.astype("float32")
X = X.toarray()
vocab = vectorizer.get_feature_names_out()
tf_idf_matrix = pd.DataFrame(np.round(X, 2), columns=vocab)
tf_idf_matrix
```

```python
similarity_matrix = cosine_similarity(X)
```

```python
# np.save("similarity_mat.npy", similarity_matrix)
# similarity_matrix = np.load("similarity_mat.npy")
```

```python
def find_duplicates(matrix):
    indices = np.argwhere(matrix == 1) # 1 cause a lot of similarities are high
    duplicates = [(i, j) for i, j in indices if i < j]
    return duplicates

duplicates_c = find_duplicates(similarity_matrix)
len(duplicates_c)
```
